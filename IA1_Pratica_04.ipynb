## A busca em largura é melhor nesse caso porque garante o menor caminho até o objetivo.

```python
grafo = {
    'A': ['B', 'C', 'D'], 
    'B': ['D'],           
    'C': ['D', 'E'],     
    'D': ['B', 'G', 'E'], 
    'E': ['F'],           
    'F': ['H'],           
    'G': ['H'],           
    'H': []               
}
print(grafo)
```

```python
#Busca em Largura
def bfs_caminho(grafo, inicio, objetivo):
    fila = [[inicio]]
    visitados = set()
    while fila:
        caminho = fila.pop(0) # BFS usa Fila [cite: 18]
        no = caminho[-1]
        if no == objetivo: return caminho
        if no not in visitados:
            for vizinho in grafo.get(no, []):
                novo_caminho = list(caminho)
                novo_caminho.append(vizinho)
                fila.append(novo_caminho)
            visitados.add(no)
    return None
print(bfs(grafo, 'A', 'H'))
```

```python
#Busca em Profundidade
def dfs(grafo, inicio, objetivo):
    pilha = [[inicio]]
    visitados = set()
    while pilha:
        caminho = pilha.pop()
        no = caminho[-1]
        if no == objetivo: return caminho
        if no not in visitados:
            for vizinho in reversed(grafo.get(no, [])):
                novo_caminho = list(caminho)
                novo_caminho.append(vizinho)
                pilha.append(novo_caminho)
            visitados.add(no)
print(dfs(grafo, 'A', 'H'))
```

## A Busca em Largura e a Busca em Profundidade apresentam resultados diferentes devido à forma como exploram o grafo. A Busca em Largura percorre os nós nível por nível, garantindo encontrar o caminho com menor número de arestas, sendo melhor quando se deseja o melhor caminho. A Busca em Profundidade explora um caminho até o máximo de profundidade antes de voltar, podendo encontrar soluções mais longas e não necessariamente ótimas.
